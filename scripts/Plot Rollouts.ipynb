```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
inp =  np.load('/media/home/alex/aihack22/data/rollouts/alex_orig_sequence.npy')
base =  np.load('/media/home/alex/aihack22/data/rollouts/alex_normal.npy')
base_x2 =  np.load('/media/home/alex/aihack22/data/rollouts/alex_normal_x2.npy')
base_x4 =  np.load('/media/home/alex/aihack22/data/rollouts/alex_normal_x4.npy')
base_x10 =  np.load('/media/home/alex/aihack22/data/rollouts/alex_normal_x10.npy')
base_x20 =  np.load('/media/home/alex/aihack22/data/rollouts/alex_normal_x20.npy')

len_plot = 10
skips = [1, 1, 2, 4, 10, 20]
labels = ['Input','Reco (step size=1)', 'Reco (step size=0.5)', 'Reco (step size=0.25)','Reco (step size=0.1)','Reco (step size=0.05)']
f, ax = plt.subplots(6, len_plot, figsize=(len_plot*5,len_plot*3))
for row, rollout in enumerate([inp, base, base_x2, base_x4, base_x10, base_x20]):
    for i, step in enumerate(range(0,skips[row]*len_plot, skips[row])):
        ax[row,i].imshow(rollout[0,step,0])
        ax[row,i].set_xticks([])
        ax[row,i].set_yticks([])
        if i ==0:
            ax[row, i].set_ylabel(labels[row], fontsize=20)
        if row == 0:
            ax[row,i].set_title(f't={step}', fontsize=20)
plt.tight_layout()
plt.savefig('/media/home/alex/aihack22/data/main_out.pdf')
```
