```python
inp =  np.load('/media/home/alex/aihack22/data/rollouts/alex_orig_sequence.npy')
base =  np.load('/media/home/alex/aihack22/data/rollouts/alex_normal.npy')
base_x2 =  np.load('/media/home/alex/aihack22/data/rollouts/alex_normal_x2.npy')
base_x4 =  np.load('/media/home/alex/aihack22/data/rollouts/alex_normal_x4.npy')
base_x10 =  np.load('/media/home/alex/aihack22/data/rollouts/alex_normal_x10.npy')
base_x20 =  np.load('/media/home/alex/aihack22/data/rollouts/alex_normal_x20.npy')

len_plot = 10



from celluloid import Camera
fig = plt.figure()
camera = Camera(fig)
# f, ax = plt.subplots(6, len_plot, figsize=(len_plot*5,len_plot*3))
datasets = [inp, base, base_x2, base_x4, base_x10, base_x20]
skips = [1, 1, 2, 4, 10, 20]
labels = ['Ground Truth','Reco (step size=1)', 'Reco (step size=0.5)', 'Reco (step size=0.25)','Reco (step size=0.1)','Reco (step size=0.05)']
manual = 2
for row, rollout in enumerate([datasets[manual]]):
    for i, step in enumerate(range(0,skips[manual]*len_plot, skips[manual])):
        plt.imshow(rollout[0,i,0])
        plt.text(2,7, str(step), c='white')
        plt.xticks([])
        plt.yticks([])
        # if i ==0:
        plt.title(str(labels[manual]), fontsize=20)
        camera.snap()
        # if row == 0:
        # ax[row,i].set_title(, fontsize=20)
animation = camera.animate()
animation.save('./animation_base.gif')
# plt.tight_layout()

# plt.savefig('/media/home/alex/aihack22/data/main_out.pdf')
```
