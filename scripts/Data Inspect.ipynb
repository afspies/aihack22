```python
import numpy as np
data = np.load('/media/home/alex/aihack22/data/trajectories_full/t1.npy')

import matplotlib.pyplot as plt
```

```python
print(np.unique(data))
```

```python
side = 6
from celluloid import Camera
fig = plt.figure()
camera = Camera(fig)
iter_Data = iter(data)
for i in range(side**2):
    batch = next(iter_Data)
    # row = i % side
    # col =  i //side 
    plt.imshow(batch)
    plt.title(str(i))
    camera.snap()
    # ax[row, col].imshow(batch, cmap='gray')
    # ax[row, col].axis('off')
```
